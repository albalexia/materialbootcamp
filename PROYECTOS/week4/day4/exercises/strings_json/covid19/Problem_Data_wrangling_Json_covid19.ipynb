## Working on Jsons

```python
import requests
import json
import pandas as pd #para no tener que escribir "pandas"todo el rato
```

```python
r = requests.get(url='https://covid19tracking.narrativa.com/feed_en.json')
json_readed = json.loads(r.text) #transformarla a textp para leerlo
json_readed  #accede a la url, es una peticion(request)
```

```python
print(json_readed)
```

Save json in a local file called "data.json"

```python
with open('data.json', 'w+') as outfile: #abre un archivo data.jason. dump vuelca los datos
    json.dump(json_readed, outfile) #vuelca los datos de json_readed en outfile
type(json_readed)
```

```python
with open('data.json') as f:  #abre el fichero y lo imprime linea a linea
    line = f.readline()
    if line:
        print(line)
```

Save with indent

```python
with open('data_indented.json', 'w+') as outfile: #abro un fichero en modo escritura 
    json.dump(json_readed, outfile, indent=4) #escribo la estructura como json
```

Read local json


```python
with open('data_indented.json', 'r+') as outfile: #abro el fichero en modo escritura
    json_readed = json.load(outfile)#load es cargar
print(type(json_readed))
json_readed
```

```python
df_json = pd.read_json("data_indented.json")
df_json
```
