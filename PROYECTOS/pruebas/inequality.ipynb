```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.style.use('fivethirtyeight')
```

```python
def read_csv_gob(f, comma_column='unit,indic_il,sex,age,geo\\time'):
    df = pd.read_csv(f, sep="\t")
    df[comma_column.split(",")]=df[comma_column].str.split(',',expand=True)
    del df[comma_column]
    return df
```
